1. Read Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_absolute_error, mean_squared_error, confusion_matrix, classification_report

from sklearn.linear_model import LogisticRegression, SGDClassifier, Perceptron
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.neural_network import MLPClassifier

# Load dataset
df = pd.read_csv("Employee.csv")
print(df.head())
print(df.shape)
print(df.info())
print(df.describe(include='all'))
```

2. Preprocessing

```python
df.drop_duplicates(inplace=True)

binary_cols = ['Gender', 'EverBenched']
for col in binary_cols:
    df[col] = LabelEncoder().fit_transform(df[col])

multi_cat_cols = ['Education', 'City']
df = pd.get_dummies(df, columns=multi_cat_cols, drop_first=True)

scaler = StandardScaler()
num_cols = ['Age', 'JoiningYear', 'PaymentTier', 'ExperienceInCurrentDomain']
df[num_cols] = scaler.fit_transform(df[num_cols])

X = df.drop('LeaveOrNot', axis=1)
y = df['LeaveOrNot']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

3. Visualization

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='LeaveOrNot', data=df)
plt.title("Target Class Distribution")
plt.tight_layout()
plt.show()
```

4. ML Models

```python
models = {
    'LogisticRegression': LogisticRegression(),
    'GaussianNB': GaussianNB(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'SGDClassifier': SGDClassifier(),
    'LinearSVC': LinearSVC(),
    'Perceptron': Perceptron(),
    'MLPClassifier': MLPClassifier()
}

model_names = []
model_scores = []

for name, model in tqdm(models.items()):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    model_names.append(name)
    model_scores.append([
        accuracy_score(y_test, y_pred),
        precision_score(y_test, y_pred),
        recall_score(y_test, y_pred),
        f1_score(y_test, y_pred),
        mean_absolute_error(y_test, y_pred),
        mean_squared_error(y_test, y_pred)
    ])
```

5. Metrics Table

```python
results_df = pd.DataFrame(model_scores, index=model_names, columns=['Accuracy', 'Precision', 'Recall', 'F1 Score', 'MAE', 'MSE'])
results_df = results_df.sort_values(by='F1 Score', ascending=False)
print("\nModel Evaluation Table:\n", results_df)
```

6. Best Model Evaluation

```python
best_model_name = results_df.index[0]
best_model = models[best_model_name]
y_pred_best = best_model.predict(X_test)

print(f"\nBest Model: {best_model_name}")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
print("\nClassification Report:\n", classification_report(y_test, y_pred_best))

plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred_best), annot=True, fmt='d', cmap='Blues')
plt.title(f"Confusion Matrix - {best_model_name}")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# Plot all metrics
results_df[['Accuracy', 'Precision', 'Recall', 'F1 Score']].plot(kind='bar', figsize=(12, 6))
plt.title('Model Accuracy Scores')
plt.xlabel('Model')
plt.ylabel('Scores')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
